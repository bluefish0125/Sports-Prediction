```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import pytz
```

```python
folder_path = "Processed/"
csv_path = "NBA_2018-2024.csv"
complete_path = folder_path + csv_path
df = pd.read_csv(complete_path, index_col=0)
df = df.reset_index(drop=True)
```

```python
mapping = {'Brooklyn Nets': 'BRK', 'Golden State Warriors': 'GSW', 'Los Angeles Lakers': 'LAL',
       'Milwaukee Bucks': 'MIL', 'Boston Celtics': 'BOS', 'Charlotte Hornets': 'CHO',
       'Chicago Bulls': 'CHI', 'Cleveland Cavaliers': 'CLE', 'Denver Nuggets': 'DEN',
       'Detroit Pistons': 'DET', 'Houston Rockets': 'HOU', 'Indiana Pacers': 'IND',
       'Memphis Grizzlies': 'MEM', 'Minnesota Timberwolves': 'MIN',
       'New Orleans Pelicans': 'NOP', 'New York Knicks': 'NYK', 'Oklahoma City Thunder': 'OKC',
       'Orlando Magic': 'ORL', 'Philadelphia 76ers': 'PHI', 'Phoenix Suns': 'PHO',
       'Portland Trail Blazers': 'POR', 'Sacramento Kings': 'SAC', 'San Antonio Spurs': 'SAS',
       'Toronto Raptors': 'TOR', 'Utah Jazz': 'UTA', 'Washington Wizards': 'WAS',
       'Atlanta Hawks': 'ATL', 'Dallas Mavericks': 'DAL', 'LA Clippers': 'LAC', 'Miami Heat': 'MIA'}
green = []
keys = []
for key in mapping.keys():
    green.append(mapping[key])
    keys.append(key)
mapping = {}
for i in range(len(green)):
    mapping[green[i]] = keys[i]
```

```python
dates = df['date'].unique()
```

```python
Teams = []
for date in dates:
    Teams.append(df[df['date'] == date]['Teams'])
```

```python
dates = df['date'].unique()
```

```python
# Define time zones
dates_utc = []
est_timezone = pytz.timezone('America/New_York')  # Eastern Standard Time (EST)
utc_timezone = pytz.timezone('UTC')

for date in dates:
    date = date[:10]
    # Create a datetime object for the given time in EST
    est_time = datetime.strptime(f'{date} 12:00 PM', '%Y-%m-%d %I:%M %p')
    est_time = est_timezone.localize(est_time)

    # Convert EST time to UTC
    utc_time = est_time.astimezone(utc_timezone)
    final_time = utc_time.strftime('%Y-%m-%dT%H:%M:%SZ')
    dates_utc.append(final_time)
```

```python
full = []
SPORT = "basketball_nba"
REGION = "us"
FORMAT = "decimal"
for x in range(24):
    timestamp_str = "2024-02-15T00:00:00"
    timestamp = datetime.fromisoformat(timestamp_str)
    time = timestamp + timedelta(minutes=5*x)
    timestamp = time.strftime('%Y-%m-%dT%H:%M:%S') + "Z"
    test = requests.get(f'https://api.the-odds-api.com/v4/historical/sports/{SPORT}/odds?regions={REGION}&oddsFormat={FORMAT}&apiKey={KEY}&date={timestamp}').json()
    for i in range(len(test['data'])):
        num_outcomes = len(test['data'][0]['bookmakers'][0]['markets'][0]['outcomes'])
        best_odds = [[None,None,None,None, float('-inf')] for _ in range(num_outcomes)]
        bookmakers = test['data'][i]['bookmakers']
        for index, data in enumerate(bookmakers):
            for outcome in range(num_outcomes):
                odds = test['data'][i]['bookmakers'][index]['markets'][0]['outcomes'][outcome]['price']
                current_odds = best_odds[outcome][4]
                if odds > current_odds: #and test['data'][i]['bookmakers'][index]['title'] != "Betfair":
                    best_odds[outcome][0] = test['timestamp']
                    best_odds[outcome][1] = test['data'][i]['id']
                    best_odds[outcome][2] = test['data'][i]['bookmakers'][index]['title']
                    best_odds[outcome][3] = test['data'][i]['bookmakers'][index]['markets'][0]['outcomes'][outcome]['name']
                    best_odds[outcome][4] = odds
        for outcome in range(num_outcomes):
            full.append(best_odds[outcome])
```

```python
timestamp = []
Id = []
sportsbook = []
team = []
odds = []
for i in range(len(full)):
    timestamp.append(full[i][0])
    Id.append(full[i][1])
    sportsbook.append(full[i][2])
    team.append(full[i][3])
    odds.append(full[i][4])
df = pd.DataFrame({"Timestamp": timestamp,"Id": Id, "Sportsbook": sportsbook, "Team": team, "Odds": odds})
df.head()
```
