### 2.1 Traditional Stats Exploration <a id='2_1'></a>

- `MIN`: Minutes per game.
- `FGM`: Field Goals Made per game.
- `FGA`: Field Goals Attempted per game.
- `FG_PCT`: Field Goal Percentage.
- `FG3M`: Three-Point Field Goals Made per game.
- `FG3A`: Three-Point Field Goals Attempted per game.
- `FG3_PCT`: Three-Point Field Goal Percentage.
- `FTM`: Free Throws Made per game.
- `FTA`: Free Throws Attempted per game.
- `FT_PCT`: Free Throw Percentage.
- `OREB`: Offensive Rebounds per game.
- `DREB`: Defensive Rebounds per game.
- `REB`: Total Rebounds per game.
- `AST`: Assists per game.
- `TOV`: Turnovers per game.
- `STL`: Steals per game.
- `BLK`: Blocks per game.
- `BLKA`: Blocks Against per game.
- `PF`: Personal Fouls per game.
- `PFD`: Personal Fouls Drawn per game.
- `PTS`: Points per game.


```python
help(leaguedashteamstats.LeagueDashTeamStats)
```

```python
# 'measure_type_detailed_defense' used for selecting for traditional or advanced
traditional_stats_jan_2022 = leaguedashteamstats.LeagueDashTeamStats(measure_type_detailed_defense='Base',
                                                         season='2021-22',
                                                         season_type_all_star='Regular Season',
                                                         date_from_nullable='01/01/2022', # Format: 'MM/DD/YYYY', data up to this point
                                                         date_to_nullable='01/31/2022' #    This is for the past month which can be used as 
                                                        ).get_data_frames()[0]         # a different set of features as recent performance
traditional_stats_jan_2022.sort_values('W_PCT', ascending=False).head()
```

```python
# For data up to a certain point which should be used for backtesting
traditional_stats_2022 = leaguedashteamstats.LeagueDashTeamStats(measure_type_detailed_defense='Base',
                                                         season='2021-22',
                                                         season_type_all_star='Regular Season',
                                                         date_to_nullable='01/31/2022' #    This is for the past month which can be used as 
                                                        ).get_data_frames()[0]         # a different set of features as recent performance
traditional_stats_2022.sort_values('W_PCT', ascending=False).head()
```

### 2.2 Advanced Stats Exploration <a id='2_2'></a>

- `OFF_RATING`: Offensive Rating - points scored per 100 possessions.
- `DEF_RATING`: Defensive Rating - points allowed per 100 possessions.
- `NET_RATING`: Net Rating - point differential per 100 possessions.
- `AST_PCT`: Assist Percentage - percentage of teammate field goals assisted.
- `AST_TO`: Assist to Turnover Ratio.
- `AST_RATIO`: Assist Ratio - assists per 100 possessions.
- `OREB_PCT`: Offensive Rebound Percentage.
- `DREB_PCT`: Defensive Rebound Percentage.
- `REB_PCT`: Rebound Percentage.
- `TM_TOV_PCT`: Turnover Percentage - turnovers per 100 plays.
- `EFG_PCT`: Effective Field Goal Percentage.
- `TS_PCT`: True Shooting Percentage - a measure of shooting efficiency.
- `USG_PCT`: Usage Percentage - percentage of team plays used by a player.
- `PACE`: Pace - estimated possessions per 48 minutes.
- `PIE`: Player Impact Estimate - a player's statistical contribution.

```python
# "date_to_nullable" will help us do training and select data up to a certain point

# Below is 2021-22
advanced_stats_2022 = leaguedashteamstats.LeagueDashTeamStats(measure_type_detailed_defense='Advanced',
                                                         season='2021-22',
                                                         season_type_all_star='Regular Season',
                                                         date_to_nullable='01/31/2022'  # Format: 'MM/DD/YYYY', data up to this point
                                                        ).get_data_frames()[0]
advanced_stats_2022.sort_values('W_PCT', ascending=False).head()
```

```python
# "date_to_nullable" will help us do training and select data up to a certain point
advanced_stats_current = leaguedashteamstats.LeagueDashTeamStats(measure_type_detailed_defense='Advanced',
                                                         season='2023-24',
                                                         season_type_all_star='Regular Season'
                                                        ).get_data_frames()[0]
# Ranked by points diffential per 100 possessions
advanced_stats_current.sort_values('NET_RATING', ascending=False).head()
```

### 2.4 Four Factors Stats Exploration <a id='2_4'></a>

- `EFG_PCT`: Effective Field Goal Percentage - Adjusts for the fact that a 3-point field goal is worth more than a 2-point field goal.
- `FTA_RATE`: Free Throw Attempt Rate - Number of free throw attempts per field goal attempt.
- `TM_TOV_PCT`: Turnover Percentage - Turnovers per 100 plays.
- `OREB_PCT`: Offensive Rebound Percentage - Percentage of available offensive rebounds obtained.
- `OPP_EFG_PCT`: Opponent Effective Field Goal Percentage.
- `OPP_FTA_RATE`: Opponent Free Throw Attempt Rate.
- `OPP_TOV_PCT`: Opponent Turnover Percentage.
- `OPP_OREB_PCT`: Opponent Offensive Rebound Percentage.

```python
four_factors_2022 = leaguedashteamstats.LeagueDashTeamStats(measure_type_detailed_defense='Four Factors',
                                                         season='2021-22',
                                                         season_type_all_star='Regular Season',
                                                         date_to_nullable='01/31/2022'  # Format: 'MM/DD/YYYY', data up to this point
                                                        ).get_data_frames()[0]
four_factors_2022.sort_values("W", ascending=False).head()
```

### 2.5 Scoring Stats Exploration <a id='2_5'></a>

- `PCT_FGA_2PT`: Percentage of field goal attempts that are 2-point shots.
- `PCT_FGA_3PT`: Percentage of field goal attempts that are 3-point shots.
- `PCT_PTS_2PT`: Percentage of points that come from 2-point field goals.
- `PCT_PTS_2PT_MR`: Percentage of points that come from mid-range 2-point shots.
- `PCT_PTS_3PT`: Percentage of points that come from 3-point field goals.
- `PCT_PTS_FB`: Percentage of points that come from fast breaks.
- `PCT_PTS_FT`: Percentage of points that come from free throws.
- `PCT_PTS_OFF_TOV`: Percentage of points that come from turnovers.
- `PCT_PTS_PAINT`: Percentage of points that come from the paint.
- `PCT_AST_2PM`: Percentage of 2-point field goals that were assisted.
- `PCT_UAST_2PM`: Percentage of 2-point field goals that were unassisted.
- `PCT_AST_3PM`: Percentage of 3-point field goals that were assisted.
- `PCT_UAST_3PM`: Percentage of 3-point field goals that were unassisted.
- `PCT_AST_FGM`: Percentage of total field goals that were assisted.
- `PCT_UAST_FGM`: Percentage of total field goals that were unassisted.


```python
scoring_2022 = leaguedashteamstats.LeagueDashTeamStats(measure_type_detailed_defense='Scoring',
                                                         season='2021-22',
                                                         season_type_all_star='Regular Season',
                                                         date_to_nullable='01/31/2022'  # Format: 'MM/DD/YYYY', data up to this point
                                                        ).get_data_frames()[0]
scoring_2022.sort_values("PCT_FGA_2PT", ascending=False).head()
```
