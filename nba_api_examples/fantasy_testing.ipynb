### NBA API

```python
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.library.parameters import SeasonAll
from nba_api.stats.static import players
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

help(leaguegamefinder)
```

```python
# Give the list of superstars to search
superstars = ["Stephen Curry", "Kawhi Leonard", "LeBron James", "Russell Westbrook"]

#initialize empty lists that will put player id into
player_ids = []
for i in superstars:
    player_ids.append(players.find_players_by_full_name(i)[0]['id'])
```

```python
import numpy as np
def average_until_this_game(df, col):
    # Getting the cumulative sum
    cumsum = (df.sort_values('Rank', ascending = True)[col].to_numpy().cumsum())/(df.sort_values('Rank', ascending = True)['Rank'].to_numpy())
    return cumsum

def adding_season_average(df, col):
    return df.join(df.groupby('SEASON_ID')[col].mean().round(2), on='SEASON_ID', rsuffix='_SEASON')

def adding_game_played_of_season(df):
    df['dummy'] = 1
    return df.join(df.sort_values('Rank').groupby(['SEASON_ID'])['dummy'].cumsum(), rsuffix='_SEASON').drop('dummy', axis=1).rename({'dummy_SEASON' : 'games_played_of_this_season'}, axis=1)

def adding_season_averaged_til_day(df, col):
    temp = df.sort_values('Rank').groupby(['SEASON_ID'])[col].cumsum()/df['games_played_of_this_season']
    temp.name = col + '_SEASON_CUM'
    df[temp.name] = temp

def adding_rolling_avergaed_til_day(df, col):
    rolling = df.sort_values('Rank')[col].rolling(15).mean()
    col_name = col + '_rolling'
    df[col_name] = rolling


def adding_career_avergaed_til_day(df, col):
    career_average = np.flip(df.sort_values('Rank')[col].to_numpy().cumsum()/df.sort_values('Rank')['Rank'].to_numpy())
    col_name = col + '_career'
    df[col_name] = career_average


```

```python
#pull game logs of each player
game_logs =[]
for i in player_ids:
    game_logs.append(playergamelog.PlayerGameLog(player_id=i, season = SeasonAll.all).get_data_frames()[0])

player_data = dict(zip(superstars, game_logs))

#Enter any player from superstar list to access their game log
kawhi = player_data["Kawhi Leonard"]
```

```python
# Columns that are being used in fantasy stock price
fantasy_cols_nba_api = ['PTS','FG3M','OREB','DREB','AST','STL','BLK','TOV']
```

```python
from datetime import datetime

kawhi['Birthday'] = datetime.strptime('JUN 29,1991', '%b %d,%Y')
kawhi['GAME_DATE'] = pd.to_datetime(kawhi['GAME_DATE'], format='%b %d, %Y')
kawhi['Age'] = ((kawhi['GAME_DATE'] - kawhi['Birthday']).dt.days / 365).astype(int)
kawhi['Rank'] = range(len(kawhi), 0,-1)
```

```python
kawhi = adding_game_played_of_season(kawhi)

for i in fantasy_cols_nba_api:
    adding_season_averaged_til_day(kawhi, i)
    adding_rolling_avergaed_til_day(kawhi, i)
    adding_career_avergaed_til_day(kawhi, i)
```

```python
kawhi['season_price'] = kawhi['PTS_SEASON_CUM'] * 0.68 + kawhi['FG3M_SEASON_CUM'] + kawhi['OREB_SEASON_CUM'] + kawhi['DREB_SEASON_CUM'] * 0.5 + kawhi['AST_SEASON_CUM'] * 0.7 + kawhi['STL_SEASON_CUM'] * 2 + kawhi['BLK_SEASON_CUM'] - kawhi['TOV_SEASON_CUM'] * 3 + (25 - kawhi['Age']) * 0.4
kawhi['career_price'] = kawhi['PTS_career'] * 0.68 + kawhi['FG3M_career'] + kawhi['OREB_career'] + kawhi['DREB_career'] * 0.5 + kawhi['AST_career'] * 0.7 + kawhi['STL_career'] * 2 + kawhi['BLK_career'] - kawhi['TOV_career'] * 3 + (25 - kawhi['Age']) * 0.4
kawhi['rolling_price'] = kawhi['PTS_rolling'] * 0.68 + kawhi['FG3M_rolling'] + kawhi['OREB_rolling'] + kawhi['DREB_rolling'] * 0.5 + kawhi['AST_rolling'] * 0.7 + kawhi['STL_rolling'] * 2 + kawhi['BLK_rolling'] - kawhi['TOV_rolling'] * 3 + (25 - kawhi['Age']) * 0.4

for i in range(kawhi.shape[0]):
    if kawhi.loc[i,'games_played_of_this_season'] < 15 and kawhi.loc[i, 'SEASON_ID'] != kawhi['SEASON_ID'].min():
        kawhi.loc[i, 'season_price'] = kawhi.loc[i,'rolling_price']
kawhi['final_price'] = 0.45 * kawhi['PTS_SEASON_CUM'] + 0.35 * kawhi['PTS_career'] + 0.2 * kawhi['PTS_rolling']

for i in range(kawhi.shape[0]):
    if kawhi.loc[i, 'SEASON_ID'] == kawhi['SEASON_ID'].min() and kawhi.loc[i,'games_played_of_this_season'] < 15:
        kawhi.loc[i, 'final_price'] = kawhi.loc[i, 'career_price']
kawhi['final_price'] = 0.45 * kawhi['PTS_SEASON_CUM'] + 0.35 * kawhi['PTS_career'] + 0.2 * kawhi['PTS_rolling']

```
