## 3/18 Games Prediction

```python
import pandas as pd
import numpy as np
import sklearn
import torch.nn as nn
import sklearn.model_selection
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
import pickle
from sklearn.metrics import accuracy_score,f1_score
pd.set_option('display.max_columns', None)
```

```python
# fill user with own user
df = pd.read_csv("/Users/benjamincheng/Documents/GitHub/Sports-Betting/data/processed_data/processed_2019_2024.csv", index_col=0)
```

```python
# use list comphrension to call games from 3/18
tonight = df[df['date_next'] == '2024-03-18 00:00:00']
tonight.reset_index(drop=True)
```

```python
# Predictors chosen using sequential feature selector trained on ridge classifier
with open('/Users/benjamincheng/Documents/GitHub/Sports-Betting/ml_notebooks/factors/factors_ridge_classifier_70_features_87.3%_2019_2024.txt', 'r') as f:
    predictors = f.read()
# Ridge classifier used no hyperparameter tuning yet
with open('/Users/benjamincheng/Documents/GitHub/Sports-Betting/ml_notebooks/weights/ridge_classifier_70_features_87.3%_2019_2024.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
predictors
```

```python
predictors = predictors.split(',')[:-1]
```

```python
X = tonight[predictors]
predictions = model.predict(X)
result = pd.Series(predictions, index = X.index)
final = pd.concat([tonight['Teams_x'], result], axis=1)
final.columns = ['Teams', 'Predictions']
final
```

```python
# 3/18 games actual results
final['Actual'] = [1,0,0,1,0,1,0,1,1,0,1,0,0,1,0,1]
```
